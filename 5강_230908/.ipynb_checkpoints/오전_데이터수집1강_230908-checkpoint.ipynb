## 02.01 Legal Issues

2가지가 있다. Opt-in 데이터, Opt-out 데이터<br>
Opt-in(whitelist) : 명시적으로 동의를 받을 때에만 수집이 가능하다.<br>
Opt-out(blacklist) : 명시적으로 거부할 때에만 수집이 불가능하다.<br>
<br>
우리나라에서는 Opt-out이 기본으로 설정되어있다. 따라서 기본적으로 정보 수집이 가능하다(2023.09 현재)<br>
개인정보에 대해서는 많은 부분이 Opt-in인으로 설정되어있다.<br>
<br>
그러니 상대방 서비스에서 우리한테 어떠한 거부사항을 들이미는지 파악하는 것이 중요하다.<br>
<br>
공익적 권리 아래 공개된 것을 완전 상업적으로 이용하지 않는 경우 합법인 경우 많다.<br>
데이터베이스나 개인정보를 활용하거나, 트래픽을 과도하게 이용하는 경우(거의 DDOS처럼 됨) 등은 대부분 불법이다.<br>
이용약관/이용방침 확인할 수 있다.<br>

robots.txt 확인할 수 있다<br>

### 02.02.01 urllib

```python
from urllib import robotparser
robot = robotparser.RobotFileParser()
robot.set_url('https://www.google.com/robots.txt')
robot.read()
robot.can_fetch('*', '/search/about')
```

위에서 볼 수 있는 바와 같이, 상위 노드가 "Disallow"인경우 하위노드가 "Allow"여도 False를 반환한다.

```python
robot.set_url('https://news.naver.com/robots.txt')
robot.read()
robot.can_fetch('*', '/')
```

GET방식 요청<br>
https\://www\.google.com\/search?<br>
    q=%EC%B9%B4%EB%A6%AC%EB%82%98&<br>
    oq=%EC%B9%B4%EB%A6%AC%EB%82%98&<br>
    aqs=chrome.0.0i355i433i512j46i433i512j0i3j0i512l2j0i3j0i512j0i3j0i512l2.2959j0j7&<br>
    sourceid=chrome&<br>
    ie=UTF-8<br>

GET방식 파라미터는 \?뒤에 나오고,
각 파라미터는 \&로 구분한다. 파라미터는 키=밸류로 구성되어있다.<br>
단, 바이트로 변환된다(키워드 등)<br>
ASCII는 바이트라 문제가 없고, non-ASCII는 바이트 인코딩하는데, charset에 따라 인코딩 방식 달라진다.<br>

우리나라에서 쓰는 인코딩은 다음 등과 같다 :<br>
ISO-9301-\?(?9301은 아니고 아무튼 ISO), EUC-KR, UTF=8<br>

Application의 Cookie/Session 값
