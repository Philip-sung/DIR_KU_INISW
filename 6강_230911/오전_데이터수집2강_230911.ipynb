## 02.02 기초 함수 작성(에러처리)

```python
#Catch Response code 200(resp), 400(raise, break), 500(retry)
def download(url, retries = 3):
    resp = None
    #robots.txt 반영해주면 좋겠다(잘 안돌아가는 관계로 생략)
    
    try:
        resp = urlopen(url)
    except HTTPError as e:
        if 500 <= e.code:
            if retires > 0:
                sleep(3)
                resp = download(url, retires - 1)
            else:
                print('재방문 횟수 초과')
                
        print(e)
        
    return None if resp == None else resp.read()
```

## 02.03 함수 개선 : Request
request 객체를 이용한 Header 전송 : user-agent, method<br>

```python
#Catch Response code 200(resp), 400(raise, break), 500(retry)
#user-agent value = 내 브라우저
def download(url, method='GET', retries = 3):
    resp = None
    #robots.txt 반영해주면 좋겠다(잘 안돌아가는 관계로 생략)
    
    try:
        req = Request(url, headers={'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/116.0.0.0 Safari/537.36'},
                             method=method)
        resp = urlopen(req)
    except HTTPError as e:
        if 500 <= e.code:
            if retires > 0:
                sleep(3)
                resp = download(url, method = method, retries = retires - 1)
            else:
                print('재방문 횟수 초과')
                
        print('Request', req.headers())
        print('Response', e)
        
    return None if resp == None else resp.read()
```

## 02.06 함수로 정리(그냥 해봄)

```python
def naver_keyword_search_url(keyword, myWhere, myPage = 1):
    naverURL = 'https://search.naver.com/search.naver?nso=&page=3&qdt=-1&query=OnebyOne&qvt=-1&sm=tab_pge&start=16&where=web'
    urlParseRes = urlparse(naverURL)
    urlQueryRes = urlParseRes.query
    pqs = parse_qs(urlQueryRes)
    print(urlParseRes)
    myKeyword = keyword
    pqs['query'] = myKeyword
    pqs['where'] = myWhere
    pqs['page'] = myPage
    pqs['start'] = 1 + (myPage-2)*15
    pqs = urlencode(pqs)
    newURL = urlunparse((*urlParseRes[:4], pqs, ''))
    
    return newURL
```

```python
naver_keyword_search_url('OnebyOne', 'image')
```

```python
def daum_keyword_search_url(keyword, myWhere='fusion', myPage = 1):
    daumURL = 'https://search.daum.net/search?w=fusion&q=%EC%A7%84%EB%89%B4%EC%8A%A4&DA=PGD&p=2'
    urlParseRes = urlparse(daumURL)
    urlQueryRes = urlParseRes.query
    pqs = parse_qs(urlQueryRes)
    print(urlParseRes)
    print(urlQueryRes)
    myKeyword = keyword
    pqs['q'] = myKeyword
    pqs['w'] = myWhere
    pqs['p'] = myPage
    pqs = urlencode(pqs)
    newURL = urlunparse((*urlParseRes[:4], pqs, ''))
    
    return newURL
```

## 02.08 함수개선(requests 사용)

```python
#Catch Response code 200(resp), 400(raise, break), 500(retry)
#user-agent value = 내 브라우저
def download(url, method='GET', retries = 3):
    resp = None
    
    try:
        resp = request(method, url, headers={'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/116.0.0.0 Safari/537.36'})
        resp.raise_for_status()
        #예외처리 쓰지 않고 바로 if로 물어봐서 처리해도 된다.
    except HTTPError as e:
        if 500 <= e.response.status_code:
            if retires > 0:
                sleep(3)
                resp = download(url, method = method, retries = retires - 1)
            else:
                print('재방문 횟수 초과')
                
        print('Request', resp.request.headers)
        print('Response', e.respones.headers)
        
    return resp
```

## 02.09 함수개선 : Body에 내용 넣을 수 있도록 함

```python
#Catch Response code 200(resp), 400(raise, break), 500(retry)
#user-agent value = 내 브라우저
def download(url, params, method='GET', retries = 3):
    resp = None
    
    try:
        resp = request(method, url, 
                       params=params if( method=='GET') else {}, 
                       data=params if( method=='POST') else {}, 
                       headers={'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/116.0.0.0 Safari/537.36'})
        resp.raise_for_status()
        #예외처리 쓰지 않고 바로 if로 물어봐서 처리해도 된다.
    except HTTPError as e:
        if 500 <= e.response.status_code:
            if retires > 0:
                sleep(3)
                resp = download(url, params = params, method = method, retries = retires - 1)
            else:
                print('재방문 횟수 초과')
                
        print('Request', resp.request.headers)
        print('Response', e.respones.headers)
        
    return resp

```
